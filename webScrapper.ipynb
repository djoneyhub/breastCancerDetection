```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#-------step 1: send http request and and get response object and data--------------
res = requests.get('https://www.go4worldbusiness.com/find?searchText=corn+oil&pg_buyers=5&pg_suppliers=1&_format=html&BuyersOrSuppliers=buyers')

txt = res.text
status = res.status_code

print("text:",txt,"\n", status)
# print the result
```

```python
#------step 2: parse the texts-------------------

#gets the content and parse it based on html component
soup = BeautifulSoup(res.content, 'html.parser')

# gets the text of the <title>(...)</title>
title = soup.title.text

# get the body component
body = soup.body

#get head
head = soup.head

print(body)
```

```python
#---------------------------Step 3: Select with Beautiful Soup-------------------------------------
country = []
for element in soup.select('span.subtitle'):
  text = element.text.strip()
  country.append(text)

print("Country:",country)
```

```python
products = []
for el in soup.select('h2.h2-item-title'):
  text = el.text.strip()
  products.append(text)

print("Products:",products)
```

```python
desc = []
for ds in soup.select('div.entity-row-description-search'):
  text = ds.text.strip()
  desc.append(text)

print("Description:",desc)
```

 'Quantity Required : 1 Twenty-Foot Container',
 'Shipping Terms : CIF',
 'Destination Port : United Kingdom',
 'Payment Terms : LC',
 'Looking for suppliers from : Worldwide',
 'Contact : Sabbir Ahmed']


```python
# Initialize separate lists for each item
proname = []
style = []
quantity = []
quantity_required = []
shipping_terms = []
destination_port = []
payment_terms = []
looking_for_suppliers = []
contact = []
```

```python
# Define a list of keys to identify relevant information
keys = [
    "Product Name",
    "Style",
    "Quantity",
    "Quantity Required",
    "Shipping Terms",
    "Destination Port",
    "Payment Terms",
    "Looking for suppliers from",
    "Contact"
]
```

```python
# Iterate through the descriptions
for text in desc:
    # Split by line breaks
    items = text.split('\r\n')

    # Create a dictionary to store key-value pairs
    item_dict = {key: None for key in keys}

    for item in items:
        for key in keys:
            if item.startswith(key):
                item_dict[key] = item.replace(key + " : ", "")

    # Append the values to their respective lists
    proname.append(item_dict["Product Name"])
    style.append(item_dict["Style"])
    quantity.append(item_dict["Quantity"])
    quantity_required.append(item_dict["Quantity Required"])
    shipping_terms.append(item_dict["Shipping Terms"])
    destination_port.append(item_dict["Destination Port"])
    payment_terms.append(item_dict["Payment Terms"])
    looking_for_suppliers.append(item_dict["Looking for suppliers from"])
    contact.append(item_dict["Contact"])
```

```python
# Print the contents of each list
print("Product Name:", proname)
print("Style:", style)
print("Quantity:", quantity)
print("Quantity Required:", quantity_required)
print("Shipping Terms:", shipping_terms)
print("Destination Port:", destination_port)
print("Payment Terms:", payment_terms)
print("Looking for suppliers:", looking_for_suppliers)
print("Contact:", contact)
```

```python
data = list(zip(country,products,proname,style,quantity,quantity_required,shipping_terms,destination_port,payment_terms,looking_for_suppliers,contact))
columns = ["Country", "Products", "Product Name", "Style", "Quantity", "Quantity Required", "Shipping Terms", "Destination Port", "Payment Terms", "Looking for Suppliers", "Contact"]
# Create a DataFrame from the list of tuples and column names
df = pd.DataFrame(data, columns=columns)
# Print the DataFrame
print(df)
```

```python
import re
```

```python
# Remove 'Wanted' keyword from the 'Products' column
df['Products'] = df['Products'].str.replace(r'^Wanted\s*:\s*', '', regex=True)
df['Product Name'] = df['Product Name'].str.replace(r'^Product Name\s*:\s*', '', regex=True)
df['Shipping Terms'] = df['Shipping Terms'].str.replace(r'^Shipping Terms\s*:\s*', '', regex=True)
df['Country'] = df['Country'].str.replace(r'Buyer From\n', '', regex=True)
df['Country'] = df['Country'].apply(lambda x: re.sub(' +', ' ', x))
```
